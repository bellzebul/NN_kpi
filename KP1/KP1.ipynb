## Контрольні питання

1. Python - це інтерпретована мова програмування. Основні відмінності включають в себе використання відступів для структури коду, простий і читабельний синтаксис, динамічну типізацію, багато бібліотек і фреймворків, мультипарадигмальність, платформено-незалежність та велику спільноту підтримки.

    Стосовно відступів в Python, це важлива частина синтаксису мови. В Python використовується відступ (зазвичай 4 пробіли або один таб) для визначення блоків коду, таких як цикли, функції, умовні оператори і т.д. Отже, відступи визначають структуру програми, і неправильні відступи можуть призвести до синтаксичних помилок.

2. Цикли в Python - це конструкції, які дозволяють виконувати одні й ті ж дії або блоки коду кілька разів. Основною різницею між циклами for та while є умова, за якою вони виконуються:

    Цикл for:
    - Використовується, коли вам заздалегідь відомо, скільки разів потрібно повторити певний блок коду або коли вам потрібно перебрати елементи в ітерованому об'єкті (наприклад, списку або рядку).
    - Використовується ітерація по послідовності або ітерабельному об'єкту.
    - Вимагає вказати діапазон або об'єкт для ітерації.
   
    Цикл while:
    - Використовується, коли умова виконання циклу може змінюватися під час виконання програми, і ви хочете, щоб цикл тривав, поки ця умова виконується. Вона перевіряється перед кожною ітерацією. Поки умова є істинною, цикл продовжується.
    - Використовується, коли не відомо заздалегідь, скільки разів потрібно повторити код.


3. Умовні конструкції в Python - це спосіб управління виконанням коду на основі певних умов. Вони використовуються для прийняття рішень в програмі, залежно від того, чи виконується певна умова. Ось як вони працюють:
   
__if__: 
Конструкція if використовується для перевірки певної умови. Якщо умова є істинною (True), то відповідний блок коду виконується.

Приклад:


```python
x = 10
if x > 5:
    print("x більше за 5")
```

У цьому прикладі, якщо значення x більше 5, то рядок "x більше за 5" буде виведено на екран.

__else__: Конструкція else використовується разом з if і дозволяє визначити блок коду, який виконується, якщо умова в if є хибною (False).

Приклад:

```python
x = 3
if x > 5:
    print("x більше за 5")
else:
    print("x не більше за 5")
```

У цьому прикладі, оскільки значення x менше 5, виведено рядок "x не більше за 5" через конструкцію else.

__elif__: Конструкція elif (скорочено від "else if") дозволяє перевіряти декілька умов послідовно. Якщо умова в if не є істинною, то перевіряються умови в elif. Перший elif, який має істинну умову, виконується, і решта умов і else не перевіряються.

Приклад:


```python
x = 5
if x > 10:
    print("x більше за 10")
elif x > 5:
    print("x більше за 5, але не більше за 10")
else:
    print("x не більше за 5")
```

У цьому прикладі, оскільки перша умова в if хибна, перевіряється умова в elif, і виводиться "x не більше за 5", бо ця умова є істинною.


Елементи списку можна змінити, присвоївши їм нові значення за індексом. Або видаливши якісь існуючи елементи, чи навпаки додавши щось нове.

Scikit-learn (також відомий як sklearn) - це бібліотека для мови програмування Python, яка спеціалізується на машинному навчанні (Machine Learning, ML) і використовується для розв'язання різних завдань у сфері аналізу даних та машинного навчання. Scikit-learn надає широкий спектр інструментів і алгоритмів, які допомагають вирішувати різноманітні завдання, включаючи такі основні:

 - Класифікація: Scikit-learn надає алгоритми для навчання моделей, які можуть класифікувати об'єкти у категорії або класи. Це використовується для задач, де потрібно визначити, до якого класу належить об'єкт, наприклад, визнавання рукописних літер, виявлення спаму у повідомленнях тощо.

 - Регресія: Scikit-learn дозволяє побудовувати регресійні моделі для прогнозування числових значень. Це може бути корисним для прогнозування цін на нерухомість, прибутку компанії або інших числових характеристик.

 - Кластеризація: Бібліотека надає методи для кластеризації даних, коли потрібно групувати схожі об'єкти разом. Наприклад, кластеризація споживачів для рекламних цілей або сегментація ринку.

 - Вимірювання якості моделей: Scikit-learn містить інструменти для оцінки якості моделей, такі як метрики класифікації (наприклад, точність, F1-оцінка) і метрики регресії (наприклад, середньоквадратична помилка).

 - Зменшення розмірності даних: Бібліотека має методи для зменшення розмірності даних, такі як метод головних компонент (PCA), що дозволяють зменшити кількість ознак та поліпшити ефективність моделей.

 - Вибірка ознак: Scikit-learn надає інструменти для вибору найбільш важливих ознак у наборі даних.

 - Навчання без учителя: Він підтримує завдання навчання без учителя, такі як аналіз головних компонент, кластеризація, виявлення аномалій та інші.

15. Scikit-learn містить декілька вбудованих наборів даних, які можна використовувати для навчання та тестування моделей машинного навчання.

Ось деякі з них разом із прикладами:

__Iris Dataset__:
Це один з найвідоміших датасетів у машинному навчанні. Він містить вимірювання трьох видів ірисів (Setosa, Versicolor, Virginica) з чотирьох ознак (довжина та ширина чашелистика та пелюстки). Використовується для задач класифікації.

__Digits Dataset__:
Цей набір даних містить зображення рукописних цифр (0-9), які були розміром 8x8 пікселів. Використовується для завдань класифікації та розпізнавання рукописних цифр.

__Wine Dataset__:
Цей набір даних містить характеристики вин з трьох різних сортів. Використовується для задач класифікації та кластеризації.

__Breast Cancer Dataset__:
Даний набір містить ознаки раку грудей та класифікацію на доброякісні та злоякісні пухлини. Використовується для задач класифікації.

__Boston Housing Dataset__:
Набір містить дані про ціни на житло в Бостоні та характеристики будинків. Використовується для завдань регресії.


16. Для завантаження вбудованого датасету з бібліотеки scikit-learn, можна використовувати модуль datasets. Цей модуль надає функції для завантаження різних вбудованих датасетів. Ось загальний підхід до завантаження та отримання даних та цільових міток:

Імпортувати необхідний модуль:

Для початку, імпортуйте модуль datasets з бібліотеки scikit-learn:

```python
from sklearn import datasets
```
Вибрати датасет та завантажити його:

Оберіть вбудований датасет за допомогою відповідної функції. Наприклад, для завантаження Iris Dataset використовуйте load_iris():

```python
iris = datasets.load_iris()
```

Для інших датасетів використовуйте аналогічні функції, такі як load_digits(), load_wine(), тощо.

Отримати доступ до даних та цільових міток:

Після завантаження датасету, ви можете отримати доступ до його даних та цільових міток наступним чином:

Для отримання даних користуйтесь атрибутом data:

```python
X = iris.data  # Отримати дані
```
Для отримання цільових міток (якщо вони є) використовуйте атрибут target:

```python
y = iris.target  # Отримати цільові мітки
```
Якщо вас цікавлять назви ознак (стовпців), ви можете отримати їх за допомогою атрибуту feature_names:

```python
feature_names = iris.feature_names
```

__Лінійний графік (Line Plot)__: Використовується для відображення залежності між змінними часу або послідовностями даних.

__Стовпчаста діаграма (Bar Chart)__: Використовується для порівняння кількісних значень між кількома категоріями.

__Гістограма (Histogram)__: Використовується для візуалізації розподілу однієї змінної.

__Box-Plot__: Допомагає візуалізувати статистичні характеристики розподілу даних, такі як медіана, квартилі та викиди.

__Кругова діаграма (Pie Chart)__: Використовується для відображення частки кожної категорії у відсотках від загальної суми.

__Scatter Plot (Точкова діаграма)__: Використовується для відображення залежності між двома змінними. Кожна точка на графіку представляє один об'єкт з даними.

__Heatmap (Теплокарта)__: Використовується для візуалізації матриці даних, де кольори відображають значення в матриці.

__3D-графік (3D Plot)__: Дозволяє візуалізувати дані в тривимірному просторі, корисний для аналізу тривимірних даних.

__Контурний графік (Contour Plot)__: Використовується для відображення залежності між трьома змінними, де лінії рівня показують однакові значення функції.

__Поверхневий графік (Surface Plot)__: Використовується для візуалізації тривимірних даних на поверхні.

__Розподільний графік (Scatter Matrix)__: Дозволяє створювати матрицю точкових графіків для аналізу залежностей між всіма парами змінних у даних.

__Гістограма градації кольорів (Color Histogram)__: Використовується для аналізу розподілу кольорів у зображеннях.

__Sankey Diagram (Санкі-діаграма)__: Використовується для відображення потоку ресурсів або інформації у системі або процесі.

Середовище для графіків: Matplotlib також надає інтерактивні середовища для створення складних візуалізацій та динамічних графіків.
